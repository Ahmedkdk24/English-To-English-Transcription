```python
from pydub import AudioSegment
audio = AudioSegment.from_mp3("/kaggle/input/kjkjdnfdjkd/AudioCutter_FGD 8-Expat Asian Female -AUH  -30-35(2).mp3")
audio = audio.set_channels(1).set_frame_rate(16000)
audio.export("/kaggle/working/audio_fixed.wav", format="wav")
```

```python
from faster_whisper import WhisperModel
import os
import librosa
import numpy as np

# Optional: Set Hugging Face token if model is gated
# os.environ["HF_TOKEN"] = "your_hf_token"

# === Load Whisper large model ===
# You can also use "large-v3" if installed locally or via huggingface
model = WhisperModel("large", device="cuda")  # or float32 if needed

# === Load and resample audio ===
audio_file = "/kaggle/working/audio_fixed.wav"
audio_data, orig_sr = librosa.load(audio_file, sr=None)
audio_data = librosa.resample(audio_data, orig_sr=orig_sr, target_sr=16000)
sample_rate = 16000

# === Smart chunking based on silence ===
intervals = librosa.effects.split(audio_data, top_db=30)  # silence threshold

# === Transcribe each detected chunk ===
all_text = []

for idx, (start, end) in enumerate(intervals):
    chunk = audio_data[start:end]
    if len(chunk) == 0:
        continue

    print(f"ğŸ”¹ Transcribing chunk {idx+1}/{len(intervals)} ({(end - start)/sample_rate:.2f}s)")

    try:
        # Run Whisper transcription
        segments, _ = model.transcribe(chunk, language="en")  # force English
        chunk_text = "".join([seg.text for seg in segments])
        all_text.append(chunk_text)
    except Exception as e:
        print(f"âš ï¸ Error transcribing chunk {idx+1}: {e}")

# === Final output ===
final_text = " ".join(all_text)

# === Save to file ===
output_filename = "whisper_large_transcript.txt"
with open(output_filename, "w", encoding="utf-8") as f:
    f.write(final_text)

print(f"\nâœ… Transcription complete. Output saved to: {output_filename}")
```
